```python
from keras.applications import MobileNetV2
from keras.utils import load_img, img_to_array
import numpy as np
from keras.applications.mobilenet_v2 import decode_predictions, preprocess_input
```

```python
model = MobileNetV2(weights='imagenet')
```

```python
def inference(x):
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    preds = model.predict(x)
    return decode_predictions(preds, top=3)[0][0][1]
```

```python
img_path = 'Cat03.jpg'
img = load_img(img_path, target_size=(224,224))
x = img_to_array(img)
inference(x)
```

```python
import cv2
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (500, 500))
    predicted = inference(frame[:,:,::-1]) #cv2 read images BGR :)
    cv2.putText(frame, predicted, (10, 20), cv2.FONT_HERSHEY_SIMPLEX,1, (255,255,255),3)
    cv2.imshow('Webcam', frame)
    if cv2.waitKey(1) == 13:
        break
cap.release()
cv2.destroyAllWindows()
```
